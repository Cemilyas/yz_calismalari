```python
import requests
from bs4 import BeautifulSoup

url = 'https://alicetinkaya.site/'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    links = soup.find_all('a')

    for link in links:
        href = link.get('href')
        text = link.text.strip()
        if href:
            print("Bağlantı metni: ", text, "URL:", href)
    print("Toplam bulunan bilgi sayısı: ", len(links), " adet")
else:
    print("Bağlantı başarısız: ", response.status_code)
```

```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
klasor_yolu = '/content/drive/My Drive/YAPAY_ZEKA_DERSI/resimler'
```

```python
resim_formatlari = ('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')
```

```python
islenen_bilgi = [f for f in os.listdir(klasor_yolu) if f.lower().endswith(resim_formatlari)]

print(f"Klasörde ", len(islenen_bilgi), "tane resim var.")

print("Resim dosyalarının isimleri ve dosya uzantıları")

for i in islenen_bilgi:
    print(i)
```

```python
image = cv2.imread('resimler/siberakademi.png')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Yüklenen Görüntünün Gösterimi")
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlanmış Görüntü")
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanıklaştırılmış Görüntü")
plt.show()
```

```python
edges = cv2.Canny(blurred_image, threshold1=100, threshold2=200)

plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama Sonucu")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Orijinal Görüntü")

plt.subplot(1, 4, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlama")

plt.subplot(1, 4, 3)
plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanık Görüntü")

plt.subplot(1, 4, 4)
plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama")

plt.show()
```
